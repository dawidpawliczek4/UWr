```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture


def run_clustering_npy(file_path):
    # === LOAD DATA ===
    X = np.load(file_path)

    # === KMeans ===
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans_labels = kmeans.fit_predict(X)
    kmeans_centers = kmeans.cluster_centers_

    # === GMM ===
    gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
    gmm_labels = gmm.fit_predict(X)
    gmm_means = gmm.means_

    # === PLOTS ===
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # KMeans result
    axs[0].scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=10, cmap='viridis')
    axs[0].scatter(kmeans_centers[:, 0], kmeans_centers[:, 1],
                   c='red', s=80, marker='X')
    axs[0].set_title("KMeans (k=3)")

    # GMM result
    axs[1].scatter(X[:, 0], X[:, 1], c=gmm_labels, s=10, cmap='viridis')
    axs[1].scatter(gmm_means[:, 0], gmm_means[:, 1],
                   c='red', s=80, marker='X')
    axs[1].set_title("GaussianMixture (k=3)")

    plt.suptitle(f"Clustering results for: {file_path}", fontsize=14)
    plt.tight_layout()
    plt.show()


run_clustering_npy("./data/data1.npy")
run_clustering_npy("./data/data2.npy")
run_clustering_npy("./data/data3.npy")
run_clustering_npy("./data/data4.npy")
run_clustering_npy("./data/data5.npy")
```
